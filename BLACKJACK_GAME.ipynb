```python
import random
```

```python
#creating the deck of cards
card_categories = ['Hearts','Diamonds','Clubs','Spades']
cards_list = ['Ace','2','3','4','5','6','7','8','9','10','Jack','Queen','King']
deck = [(card,category) for category in card_categories for card in cards_list]
```

```python
#function to assign points for the specific tasks
def card_value(card):
  if card[0] in ['Jack','Queen','King']:
    return 10
  elif card[0] == 'Ace':
    return 11
  else:
    return int(card[0])
```

```python
#we will use random.shuffle() to shuffle our deck of cards to maintain randomness
random.shuffle(deck)
player_card = [deck.pop(),deck.pop()]
dealer_card = [deck.pop(),deck.pop()]
```

```python
#we will be using this section to create a logic in the program
while True:
  player_score = sum(card_value(card) for card in player_card)
  dealer_score = sum(card_value(card) for card in dealer_card)
  print("Cards Player Has:",player_card)
  print("Cards Dealer Has:",dealer_card)
  print("\n")
  choice = input("Do you want to Hit or Stand? ").lower()
  if choice == "hit":
    new_card = deck.pop()
    player_card.append(new_card)
  elif choice == "stand":
    break
  else:
    print("Invalid choice. Please choose Hit or Stand.")
    continue

  if player_score > 21:
   print("Cards Dealer Has:",dealer_card)
   print("Score of the Dealer",dealer_score)
   print("Cards Player Has:",player_card)
   print("Score of the Player",player_score)
   print("Dealer Wins(Player Lost Because Player Score is exceeded 21)")
   break
```

```python
#dealer has to draw cards now
while dealer_score < 17:
    new_card = deck.pop()
    dealer_card.append(new_card)
    dealer_score += card_value(new_card)

print("Cards Dealer Has:", dealer_card)
print("Score Of The Dealer:", dealer_score)
print("\n")
```

```python
def show():
    print(f"Dealer: {dealer_card} | Score: {dealer_score}")
    print(f"Player: {player_card} | Score: {player_score}")

show()

if player_score > 21:
    print("Dealer wins (Player lost)")
elif dealer_score > 21:
    print("Player wins (Dealer lost)")
elif player_score > dealer_score:
    print("Player wins")
elif dealer_score > player_score:
    print("Dealer wins")
else:
    print("Tie")
```
